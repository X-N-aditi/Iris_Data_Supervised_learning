* perform labelencoding for Species

* import library for lebelencoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
iris['Species_encoded'] = label_encoder.fit_transform(iris['Species'])
iris[['Species_encoded']].value_counts()
```

* Now, see again head of new dataset(label_encoded dataset)

```python
iris.head()
```

* Dropping unncessary columns from dataset
* Segregate independent and dependent variable
* Check shape of new segregated variables

```python
X = iris.drop(['Species','Species_encoded'], axis = 1)
y = iris['Species_encoded']
```

```python
print(X.shape)
print(y.shape)
```

* see head of independent variables

```python
X.head()
```

* see head of dependent variable

```python
Y = pd.DataFrame(y)
Y.head()
```

* Now splitting data into train and test for further analysis

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
scaled_X_train = standard_scaler.fit_transform(X_train)
scaled_X_test = standard_scaler.fit_transform(X_test)
```

```python
logis_model = LogisticRegression()
logis_model.fit(scaled_X_train, np.ravel(y_train))
```

```python
y_pred_logis = logis_model.predict(scaled_X_test)
y_pred_logis
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred_logis)
```
